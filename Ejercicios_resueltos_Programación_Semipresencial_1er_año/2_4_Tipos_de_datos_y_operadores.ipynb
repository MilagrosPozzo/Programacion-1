# Operadores logicos y de comparacion

Los operadores l칩gicos y de comparaci칩n son fundamentales en la programaci칩n, ya que permiten evaluar condiciones y realizar acciones basadas en el resultado de esas condiciones. En Python, tenemos operadores de comparaci칩n para comparar valores y operadores l칩gicos para combinar condiciones.

## Operadores de comparacion
Los operadores de comparaci칩n son utilizados para comparar dos valores y devuelven un valor booleano, es decir, True o False. Los operadores de comparaci칩n en Python incluyen:

* Igual (==)
* Diferente (!=)
* Mayor que (>)
* Menor que (<)
* Mayor o igual que (>=)
* Menor o igual que (<=)

Ejemplo de uso de operadores de comparaci칩n:

```python
a = 5
b = 7

print("a == b:", a == b) # Falso, porque 5 no es igual a 7
print("a != b:", a != b) # Verdadero, porque 5 es diferente de 7
print("a > b:", a > b)   # Falso, porque 5 no es mayor que 7
print("a < b:", a < b)   # Verdadero, porque 5 es menor que 7
print("a >= b:", a >= b) # Falso, porque 5 no es mayor o igual que 7
print("a <= b:", a <= b) # Verdadero, porque 5 es menor o igual que 7
```

## Operadores logicos
Los operadores l칩gicos son utilizados para combinar condiciones. Los operadores l칩gicos en Python incluyen:

* and: Devuelve True si ambas condiciones son verdaderas, de lo contrario devuelve False.
* or: Devuelve True si al menos una de las condiciones es verdadera, de lo contrario devuelve False.
* not: Devuelve True si la condici칩n es falsa, y False si la condici칩n es verdadera.

Ejemplo de uso de operadores l칩gicos:

```python
x = 5
y = 10
z = 15

# Comprobar si x es menor que y y si y es menor que z
condicion1 = x < y and y < z
print("Condici칩n 1 (x < y y y < z):", condicion1) # Verdadero

# Comprobar si x es igual a y o si y es igual a z
condicion2 = x == y or y == z
print("Condici칩n 2 (x == y o y == z):", condicion2) # Falso

# Negar la condici칩n 1
condicion3 = not condicion1
print("Condici칩n 3 (negaci칩n de la condici칩n 1):", condicion3) # Falso
```

En este ejemplo, hemos utilizado operadores de comparaci칩n y operadores l칩gicos para construir condiciones m치s complejas y realizar acciones basadas en el resultado de esas condiciones.
